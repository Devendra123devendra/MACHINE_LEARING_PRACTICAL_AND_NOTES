# Linear regression practical implimantation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import fetch_openml

boston = fetch_openml(name='boston', version=1, as_frame=True, parser='pandas')


```

```python
boston
```

```python
df = boston.frame
```

```python
df.head()
```

```python
df['price'] = boston.target
df.head()
```

# dividing the dataset into dependent and independant feature

```python
x = df.iloc[:,:-1]
```

```python
x.head()
```

```python
y = df.iloc[:,-1]
```

```python
y.head()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

# convert to float
x = x.astype(float)
y = y.astype(float)

# model create
lin_reg = LinearRegression()

# cross validation with MSE
mse = cross_val_score(lin_reg, x, y, scoring='neg_mean_squared_error', cv=5)

print(mse)

```

```python


mean_mse = np.mean(mse)

print('MSE score per fold:', mse)
print('Avg Negative MSE:', mean_mse)


```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 42)
```

```python
lin_reg.fit(x_train,y_train)
```

```python
prediction = lin_reg.predict(x_test)
```

```python
print(prediction)
```

```python
comparision_df = pd.DataFrame({'Actual': y_test,'predicted':prediction})
print(comparision_df.head())
```

```python
from sklearn import metrics


# Calculate metrics
r2 = metrics.r2_score(y_test, prediction)
mae = metrics.mean_absolute_error(y_test, prediction)
mse = metrics.mean_squared_error(y_test, prediction)
rmse = np.sqrt(mse)

# Print results with formatting
print(f"R2 Score: {r2:.4f}")
print(f"MAE (Mean Absolute Error): {mae:.4f}")
print(f"MSE (Mean Squared Error): {mse:.4f}")
print(f"RMSE (Root Mean Squared Error): {rmse:.4f}")

```

# ridge regression 

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()
```

```python

# Correct alpha values
params = {'alpha': [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20]}

# GridSearchCV
ridge_regressor = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv=5)

# Train model
ridge_regressor.fit(x, y)

```

```python

# Best alpha value
print("Best alpha:", ridge_regressor.best_params_)

# Best score
print("Best score:", ridge_regressor.best_score_)
```

```python
import numpy as np

from sklearn.linear_model import LinearRegression


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
x = x.astype(float)
y = y.astype(float)

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Split dataset (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)


```

```python
model = LinearRegression()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
lin_pred = model.predict(x_test)

lin_acc = r2_score(y_test, lin_pred) * 100

print(f'Linear Regression Accuracy is : {lin_acc:.2f}%')
```

# ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

# Create Ridge model
ridge_reg = Ridge(alpha=10.0)

# Train model on training data
ridge_reg.fit(x_train, y_train)

# Predict on test data
ridge_pred = ridge_reg.predict(x_test)

# Calculate accuracy
ridge_acc = r2_score(y_test, ridge_pred) * 100

print(f'Ridge Regression Accuracy is : {ridge_acc:.2f}%')
```

# lasso regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score

# Create Ridge model
lasso_reg = Lasso(alpha=10.0)

# Train model on training data
lasso_reg.fit(x_train, y_train)

# Predict on test data
lasso_pred = lasso_reg.predict(x_test)

# Calculate accuracy
lasso_acc = r2_score(y_test, lasso_pred) * 100

print(f'Lasso Regression Accuracy is : {lasso_acc:.2f}%')
```
