# heirachical clustering

# import the library

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
```

# drow the dandrogram 

```python
x,_ = make_blobs(n_samples=50,centers=3,cluster_std=0.6,random_state=42)
plt.figure(figsize=(10,5))
dendrogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title("Dendrogram (tree diagram)")
plt.xlabel('Data Point')
plt.ylabel('Euclidean Distance')
plt.show()
```

# train the model 

```python
model = AgglomerativeClustering(n_clusters=3,linkage='ward')
y_pred = model.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1], c = y_pred,cmap= 'rainbow',s=50)
plt.title('Hierarchical Clustering Result')
plt.show()
```

# DB Scan Algoritham

```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
x,_ = make_moons(n_samples=250,noise=0.05,random_state=42)
```

```python
dbscan = DBSCAN(eps=0.2,min_samples=5)
```

```python
y_pred = dbscan.fit_predict(x)
```

# visualization result


```python
plt.figure(figsize=(8,5))
plt.scatter(x[:,0],x[:,1], c = y_pred,cmap= 'plasma',s=40)
plt.title("DBSCAN Clustering: Handling Non-Linear Shape ")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
